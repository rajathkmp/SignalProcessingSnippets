```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(-5,5,0.01)
N = 5 # Odd N
z = (np.sin(np.pi*N*t))/(N*np.sin(np.pi*t))

plt.plot(t, z)
plt.ylim(-1.5,1.5)
plt.xlim(-5,5)
plt.grid()
```

```python
N = 4 # Even N
z = (np.sin(np.pi*N*t))/(N*np.sin(np.pi*t))

plt.plot(t, z)
plt.ylim(-1.5,1.5)
plt.xlim(-5,5)
plt.grid()
```
